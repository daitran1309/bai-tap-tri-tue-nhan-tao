```python


# Trạng thái đích cần đạt được
GOAL = [1, 2, 3,
        8, 0, 4,
        7, 6, 5]   # 0 là ô trống

# ============================================================
# BƯỚC 1: Định nghĩa NODE
# Mỗi node lưu: trạng thái hiện tại + đường đi từ đầu đến đây
# ============================================================

def tao_node(state, duong_di):
    node = {
        "state": state,        # mảng 9 phần tử, ví dụ [1,2,3,4,0,6,7,5,8]
        "duong_di": duong_di   # danh sách các state từ đầu đến node này
    }
    return node


# ============================================================
# BƯỚC 2: Tìm các node con (children) của một node
# Di chuyển ô trống lên/xuống/trái/phải
# ============================================================

def tim_cac_node_con(node):
    state = node["state"]
    vi_tri_trong = state.index(0)   # tìm vị trí ô trống

    # Ô trống đang ở hàng nào, cột nào?
    hang = vi_tri_trong // 3
    cot  = vi_tri_trong % 3

    cac_node_con = []

    # Thử 4 hướng di chuyển: lên, xuống, trái, phải
    huong = [
        ("lên",    hang - 1, cot),
        ("xuống",  hang + 1, cot),
        ("trái",   hang,     cot - 1),
        ("phải",   hang,     cot + 1),
    ]

    for ten_huong, hang_moi, cot_moi in huong:
        # Kiểm tra có ra ngoài bảng không
        if 0 <= hang_moi < 3 and 0 <= cot_moi < 3:
            vi_tri_moi = hang_moi * 3 + cot_moi

            # Tạo state mới bằng cách đổi chỗ ô trống và ô bên cạnh
            state_moi = state.copy()
            state_moi[vi_tri_trong] = state_moi[vi_tri_moi]
            state_moi[vi_tri_moi]  = 0

            # Tạo node con
            duong_di_moi = node["duong_di"] + [state_moi]
            node_con = tao_node(state_moi, duong_di_moi)
            cac_node_con.append(node_con)

    return cac_node_con


# ============================================================
# BƯỚC 3: In bảng 3x3 ra màn hình cho dễ nhìn
# ============================================================

def in_bang(state):
    for i in range(3):
        hang = state[i*3 : i*3+3]
        hang_str = []
        for so in hang:
            if so == 0:
                hang_str.append(" _")
            else:
                hang_str.append(f" {so}")
        print(" ".join(hang_str))
    print()


# ============================================================
# BƯỚC 4: Thuật toán BFS-Graph (đúng như slide)
# ============================================================

def BFS_Graph(trang_thai_dau, trang_thai_dich):

    print("=" * 40)
    print("BẮT ĐẦU BFS-Graph")
    print("=" * 40)

    # --- node ← NODE(problem.INITIAL) ---
    node = tao_node(trang_thai_dau, [trang_thai_dau])

    # --- if GOAL-TEST(node.STATE) then return SOLUTION ---
    if node["state"] == trang_thai_dich:
        print("Trạng thái ban đầu đã là đích rồi!")
        return node["duong_di"]

    # --- frontier ← FIFO-QUEUE() ---
    # Dùng mảng (list) làm hàng đợi FIFO
    # Thêm vào cuối (.append), lấy ra từ đầu ([0] rồi .pop(0))
    frontier = [node]

    # --- explored ← {} (tập rỗng) ---
    # Dùng list lưu các state đã khám phá
    explored = []

    buoc = 0

    # --- while not EMPTY?(frontier) do ---
    while len(frontier) > 0:

        buoc += 1

        # --- node ← frontier.REMOVE() (lấy node đầu tiên trong queue) ---
        node = frontier[0]
        frontier.pop(0)

        # --- explored ← explored ∪ {node.STATE} ---
        explored.append(node["state"])

        print(f"Bước {buoc}: Đang xét state | frontier={len(frontier)} | explored={len(explored)}")

        # --- for each action in problem.ACTIONS(node.STATE) do ---
        cac_con = tim_cac_node_con(node)

        for child in cac_con:

            # --- if child.STATE ∉ explored AND child ∉ frontier then ---
            da_explored = child["state"] in explored
            da_frontier = any(f["state"] == child["state"] for f in frontier)

            if not da_explored and not da_frontier:

                # --- if GOAL-TEST(child.STATE) then return SOLUTION ---
                if child["state"] == trang_thai_dich:
                    print(f"\nTÌM THẤY LỜI GIẢI sau {buoc} bước!")
                    print(f"Số nodes đã explored: {len(explored)}")
                    print(f"Độ dài đường đi: {len(child['duong_di']) - 1} bước di chuyển\n")
                    return child["duong_di"]

                # --- frontier.INSERT(child) ---
                frontier.append(child)

    # --- return failure ---
    print("THẤT BẠI: Không tìm thấy lời giải!")
    return None


# ============================================================
# CHẠY CHƯƠNG TRÌNH
# ============================================================

if __name__ == "__main__":

    # Trạng thái ban đầu (thay đổi tùy ý)
    trang_thai_dau = [2, 8, 3,
                      1, 6, 4,
                      7, 0, 5]

    print("Trạng thái ban đầu:")
    in_bang(trang_thai_dau)

    print("Trạng thái đích:")
    in_bang(GOAL)

    # Chạy BFS
    loi_giai = BFS_Graph(trang_thai_dau, GOAL)

    # In lời giải từng bước
    if loi_giai:
        print("=" * 40)
        print("LỜI GIẢI TỪNG BƯỚC:")
        print("=" * 40)
        for i, state in enumerate(loi_giai):
            if i == 0:
                print(f"Bước {i} (ban đầu):")
            else:
                print(f"Bước {i}:")
            in_bang(state)
```
